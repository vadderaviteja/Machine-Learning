#### Assumptions of multiple linear regresssion:
#### Linearity:
The relationship betweeen dependent and independent variables dhould be linear.
#### Lack of multicollinearity:

It is assumed that there is little or no multicollinearity in the data. Muticollinearity occurs when the features( or independent variables )are not independet of each other.
#### Multivariate normality:
Multiple regression assumes thath the resodued are normally distributed.





"A dummy variable is a binary variable (0 or 1) used to represent categorical data in numerical form. Since most machine learning algorithms and statistical models work only with numerical input, we convert categorical variables into dummy variables.

For example, if we have a column called 'City' with categories like 'Delhi', 'Mumbai', and 'Hyderabad', we can create separate dummy variables like 'City_Delhi', 'City_Mumbai', etc. Each of them will hold a value of 1 if the category is present, and 0 otherwise.

This technique is commonly known as 'one-hot encoding'. In regression models, we often drop one dummy variable to avoid the dummy variable trap, which causes multicollinearity."

### R (R-squared) 
It is a performance metric used to evaluate how well a regression model fits the data.

It tells you what percentage of the variance in the dependent variable (target) is explained by the independent variables (features) in your model.



```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Windows\ML_Theory\50_Startups_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encode', OneHotEncoder(), [3])], remainder='passthrough')
X=ct.fit_transform(x)
# Apply transformations
```

```python
X
```

```python
X=X[:, 1:]    # avoiding the dummy variable trap ,  we are avoiding the oth column
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(X, y , test_size=0.2, random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x_train, y_train)
```

```python
y_pred=rg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
