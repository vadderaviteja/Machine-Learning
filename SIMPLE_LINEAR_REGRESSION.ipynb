# Simple Linear Regression
A simple linear regresssion model only has two explanatory variables- oone dependent and one independent variable. Think of two dimensional space where the horizontal axis represents the independent variable x, and the verical axis represents the independent variable y.

### Assumptions of simple linear regresssion
#### Linearity: 
It assumes that the realtion between the data points is linera that is plot of the data points shows is linear.
#### Independence:
There must be no relationship among the different values of the independent variable that value is unique.
#### Normality:
Data points follow a normal distribution that is if we plot the histogram of the data. we should be able to draw a skewed line.

```python
x=df.iloc[:, :1].values
y=df.iloc[:, -1].values
```

```python
# plot the before predicting
plt.plot(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, y_train, x_test, y_test=train_test_split(x, y, test_size=0.5, random_state=0)
```

```python
plt.scatter(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np


# Initialize the scaler
scaler = StandardScaler()

# Fit and transform
x_train = scaler.fit_transform(x_train)
y_train=scaler.transform(y_train)

print(x_train)
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x_train, y_train)
```

```python
# Reshape x_test to 2D
x_test = x_test.reshape(-1, 1)

y_pred=rg.predict(x_train)
```

```python
import matplotlib.pyplot as plt
# Scatter plot of training data
plt.scatter(x_train, y_train, color='red', label='Training data')

# Plot the regression line
plt.plot(x_train, rg.predict(x_train), color='blue', label='Regression line')

# Labels and title
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test,y_pred, color='blue' )
plt.show()
```

```python
from sklearn.metrics import mean_squared_error


# Actual and predicted values
y_true = [3, 7, 10]
y_pred = [2.5, 7.5, 9.8]

# Calculate MSE
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import r2_score

# Actual and predicted values
y_true = [3, 7, 10, 15, 20]
y_pred = [2.8, 7.2, 9.5, 14.8, 19.5]

# Calculate R-squared
r2 = r2_score(y_true, y_pred)
print("R-Squared:", r2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample dataset (Independent variable X, Dependent variable Y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshaped for sklearn
y = np.array([2, 4, 5, 4, 5])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data points
plt.scatter(X, y, color='red', label="Actual Data")  # Scatter plot of actual data
plt.plot(X, y_pred, color='blue', label="Regression Line")  # Regression line

# Labels and title
plt.title('Simple Linear Regression')
plt.xlabel('X (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
data = np.random.normal(50, 10, 1000)  # mean=50, std=10

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='lightgreen')

# Plot bell curve using norm.pdf
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, color='blue')

# Labels
plt.title("Bell Curve (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
