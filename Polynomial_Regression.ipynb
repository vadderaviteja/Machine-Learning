# polynomial Regression

Polynomial Regression is a form of regression analysis in which the relationship between the analysis in which the relaltionship variable y is modeled as an nth degree polynomial of x. That is, if your dataset holds the characteristic of being curved when plotted in the graph, then you should go with a polynomial regression model instead of simple of Multiple Linear regresssion models.

-----------> Before choosing a regression model, we first visualize the data to understand the underlying pattern. If the relationship appears linear, we proceed with linear regression. If it shows a curve or non-linear pattern, we use polynomial regression. We compare models using performance metrics and avoid overfitting by keeping the polynomial degree optimal.



#### If the data points form a straight-line trend → Likely Linear

#### If the data points form a curve or wave → Likely Non-linear

# Import the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv(r"Position_Salaries (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
y
```

```python
x
```

#### Once observe this Below graphs, It's show  non-linear that why am using polynomial Regression.

```python
plt.plot(x,y)
```

```python
sns.pairplot(df)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=7)
x_poly=poly.fit_transform(x)
```

```python
x_poly
```

# Train the Model

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x_poly, y)
```

```python
y_poly=rg.predict(x_poly)
y_poly
```

# check the model performance with metrics

```python
from sklearn.metrics import r2_score
r2=r2_score(y_poly,y)
r2
```
